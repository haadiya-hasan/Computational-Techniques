```python
# Possible outcomes for one die
dice = [1, 2, 3, 4, 5, 6]

# Total outcomes
total_outcomes = 0
same_face_count = 0

# Generate outcomes
for d1 in dice:
    for d2 in dice:
        total_outcomes += 1
        if d1 == d2:
            same_face_count += 1

# Probability calculation
probability = same_face_count / total_outcomes

# Output
print("Total number of outcomes:", total_outcomes)
print("Number of outcomes with same faces:", same_face_count)
print("Probability of getting same faces:", probability)
```

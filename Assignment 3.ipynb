```python
#Write a python program to compute accuracy, precision, recall, F1-score for the confusion matrix

#Confusion Matrix values
TP = 199
TN = 247
FP = 3
FN = 1

#Calculations
Accuracy = (TP + TN)/(TP+TN+FP+FN)
Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
F1_Score = 2*(Precision*Recall)/(Precision+Recall)

#Outputs
print("Accuracy: ", Accuracy)
print("Precision: ", Precision)
print("Recall: ", Recall)
print("F1 - Score : ", F1_Score)
```
